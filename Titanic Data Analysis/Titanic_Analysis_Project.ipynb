###  Remove the null values from datasets  :-


```python
df.isnull().sum()
```

```python
df['Age'].value_counts()
```

```python
# check the total null values---
pd.isnull(df).sum()
```

```python
# FILL MISSING VALUES OF AGES WITH MEAN
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df['Embarked'].value_counts()
```

```python
# FILL MISSING VALUES OF Embarked with S
df['Embarked'].fillna(df['Embarked']=="S",inplace=True)
```

```python
#   drop the all missing values from datasets--

df.dropna(axis=1,inplace=True)
```

```python
#convert Age float value to int type
df['Age']=df['Age'].astype('int')
```
